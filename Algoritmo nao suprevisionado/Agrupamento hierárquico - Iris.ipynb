```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.graph_objects as go
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados = pd.read_csv("Dados\iris.csv", ",")
dados
```

```python
dados.columns = ["comprimento sépala", "largura sépala", "comprimento pétala", "largura pétala", "classe"]
dados
```

```python
dados.isnull()
```

```python
dados.drop("classe", inplace = True, axis = 1)
```

```python
dados.head()
```

```python
dados.plot.scatter(x= 'comprimento sépala', y= 'comprimento pétala')
```

```python
dados.plot.scatter(x= 'largura sépala', y= 'largura pétala')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
dados = scaler.fit_transform(dados)
dados
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

plt.figure(figsize=(20 ,15))
de = dendrogram(linkage(dados, method = "ward"))

plt.title("Iris - dendrograma")
plt.xlabel("Comprimento da sépala")
plt.ylabel("Comprimento da pétala")
```

```python
print("Valores do Dendrograma", de)
```

# Cluster

```python
from sklearn.cluster import AgglomerativeClustering

cl = AgglomerativeClustering(n_clusters = 3, affinity = "euclidean", linkage = "ward")

predict = cl.fit_predict(dados)
predict
```

```python
dados
```

```python
plt.figure(figsize=(10 ,8))

plt.scatter(dados[predict == 0, 0], dados[predict == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(dados[predict == 1, 0], dados[predict == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(dados[predict == 2, 0], dados[predict == 2, 1], s = 100, c = 'green', label = 'Cluster 3')

plt.title("Cluster Iris")
plt.xlabel('Comprimento sépala')
plt.ylabel('Comprimento pétala')
plt.legend()
```
